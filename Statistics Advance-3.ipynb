Examples of alternative hypotheses:

The average test scores of students who received tutoring are higher than the average test scores of students who did not receive tutoring.
The mean weight loss after following a particular diet plan for six weeks is greater than zero.
The proportion of voters who support Candidate A is greater than 0.50 (50%).
In hypothesis testing, the goal is to evaluate the evidence from the sample data to decide whether to reject the null hypothesis in favor of the alternative hypothesis. If there is sufficient evidence, based on the data, to reject the null hypothesis, then the alternative hypothesis is accepted, indicating that there is evidence for the claim being tested. Otherwise, if there is not enough evidence, the null hypothesis is retained.
