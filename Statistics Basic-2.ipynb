Q13. How is covariance different from correlation?

Covariance and correlation are both measures used to describe the relationship between two variables, but they have some key differences:

Scale:
Covariance is a measure of how two variables change together. It indicates the direction of the linear relationship between two variables (whether they tend to increase or decrease together), but it does not indicate the strength of the relationship.
Correlation, on the other hand, standardizes the measure of the relationship between two variables, making it easier to interpret. It measures both the strength and direction of the linear relationship between two variables and ranges between -1 and 1.
Interpretation:
Covariance can take on any value, positive, negative, or zero. A positive covariance indicates a direct relationship (as one variable increases, the other tends to increase), a negative covariance indicates an inverse relationship (as one variable increases, the other tends to decrease), and a covariance of zero indicates no linear relationship.
Correlation values range between -1 and 1. A correlation coefficient of +1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear relationship.
Units:
Covariance is expressed in the units of the variables being measured. This makes it difficult to compare covariances across different datasets or variables with different scales.
Correlation is a unitless measure, as it standardizes the covariance by dividing it by the product of the standard deviations of the two variables. This allows for easier comparison of correlation values across different datasets and variables.
In summary, while both covariance and correlation measure the relationship between two variables, correlation provides a standardized measure of this relationship, making it easier to interpret and compare across different datasets and variables.

Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

Measures of Central Tendency:
Outliers can greatly affect the mean: Since the mean is calculated by summing up all values and dividing by the total number of values, extreme values (outliers) can heavily influence the value of the mean, pulling it towards the outlier.
Outliers have minimal effect on the median: Unlike the mean, the median is not affected by extreme values. It represents the middle value when the data is ordered, so even if there are outliers, they won't impact the median as long as they don't change the order of the dataset significantly.
Outliers may or may not affect the mode, depending on their frequency and position in the dataset.
Measures of Dispersion:
Outliers can significantly affect the range: The range is the difference between the maximum and minimum values in a dataset. Outliers can greatly increase the range by pulling one end of the range further away from the other.
Outliers can increase the variance and standard deviation: Variance and standard deviation are measures of the spread of data points around the mean. Outliers can increase the variability by introducing large deviations from the mean.
Outliers may affect the interquartile range (IQR) and percentiles, depending on their position in the dataset relative to the quartiles.
Example:
Consider the following dataset representing the incomes of individuals in a small town: [20, 30, 40, 50, 60, 1000]. Here, 1000 is an outlier compared to the other incomes.

Impact on measures of central tendency:

Mean: The mean income would be greatly influenced by the outlier, pulling it towards a higher value.
Median: The median would remain relatively unaffected because it represents the middle value, which is still close to the other incomes.
Mode: The mode would remain the same since there's no change in the most frequently occurring value.
