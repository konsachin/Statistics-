Q10. In normal distribution with respect to mean position, where does the least frequent data appear?


In a normal distribution, also known as a Gaussian distribution, the least frequent data points appear farthest away from the mean, towards the tails of the distribution.

Since the normal distribution is symmetric around its mean, with the mean being the highest point of the distribution, the least frequent data points occur in the tails of the distribution, which are the regions farthest away from the mean on both sides.

Specifically, the least frequent data points are found in the tails of the distribution beyond a certain number of standard deviations from the mean. In a standard normal distribution (with a mean of 0 and a standard deviation of 1), the least frequent data points are found in the tails beyond ¬±2 or ¬±3 standard deviations from the mean.

In summary, in a normal distribution, the least frequent data points appear in the tails of the distribution, farthest away from the mean on both sides.
